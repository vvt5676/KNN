```python
from sklearn import datasets
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import math
from random import shuffle
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
def load_data_set():
    data = datasets.load_iris()
    return data['data'], data['target']
print(load_data_set()[0])
```

```python
def euclidian_distance(x_target, x_base):
    dist = 0.0
    for i in range(len(x_target)):
        dist += (x_target[i] - x_base[i])**2
    return math.sqrt(dist)
#print(len([[1, 2,3]]))
euclidian_distance([2, 3, 4], [1, 5, 6])
```

```python
def split_data(x, y, split_ratio = 0.7):
    idx_list = [i for i in range(len(x))]
    shuffle(idx_list)
    x = x[idx_list, :]
    y = y[idx_list]
    split_index = int(len(x)*split_ratio)
    x_train = x[:split_index]
    x_test = x[split_index:]
    y_train = y[:split_index]
    y_test = y[split_index:]
    return x_train, y_train, x_test, y_test
```

```python
def predict(X_test, x, y, k):
    '''
    x = np.asarray(x)
    y = np.asarray(y).reshape(-1,1)
    x = np.append(x, y,axis = 1 )
    print(np.shape(x))
    '''
    no_of_classes = len(set(y))
    print('No of classes: ', no_of_classes)
            
    results = []
    for sample in X_test:
        dist_metrics = []
        count ={}
        for i in set(y):
            count[i] =  0
        for i in range(len(x)):
            tmp_dist = euclidian_distance(sample, x[i])
            dist_metrics.append([tmp_dist]+ [y[i]])
        dist_metrics = sorted(dist_metrics, key = lambda x:x[0])
        dist_metrics = dist_metrics[:k]
        for dist_list in dist_metrics:
            count[dist_list[1]] += 1
#         print(dist_metrics)
#         print(count)
        res = max(count, key = count.get)
        results.append(res)
#         print('res: ', res)
    print(results)
    return results
# x, y = load_data_set()
# predict(predict([[1, 2, 3, 4]],x, y, 7))
```

```python
def main():
    k = 3
    x, y = load_data_set()
    
    x_train, y_train, x_test, y_test = split_data(x, y)
    print(len(x_test))
    y_predict = predict(x_test, x_train, y_train, k)
    print('without normalize: ', accuracy_score(y_predict, y_test))
    from sklearn import preprocessing
    x = preprocessing.normalize(x)
    x_train, y_train, x_test, y_test = split_data(x, y)
    y_predict = predict(x_test, x_train, y_train, k)
    print('normalized data: ', accuracy_score(y_predict, y_test), 'confusion matrix:\n',confusion_matrix(y_test, y_predict) )
    
main()
```
